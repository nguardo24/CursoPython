# Instalación de Python

Para poder empezar a escribir código en Python, el primer paso consiste en instalar el intérprete.

En tal sentido, se puede instalar Python de varias formas:

1. Página oficial
2. Instalar paquete Anaconda Distribution
3. Instalar distribución miniconda (versión minimalista de Anaconda)

## Descargar intérprete de Python desde la pagina oficial e instalar de manera local

Dirigirse a la pagína [Descargar Python](https://www.python.org/downloads/) y descargar el archivo ejecutable (.exe) para dar inicio a la instalación (notar que la ultima versión de Python a la fecha es la 3.12.2)



## Descargar e instalar Anaconda Distribution

# Instalar entorno de desarrollo integrado (Integrated Development Environment - IDE)

Una vez que tenemos `Python` instalado en el sistema, podríamos comenzar a escribir código, utilizando por ejemplo la terminal del sistema o el IDLE de Python.

Sin embargo, esta opción no es la más cómoda ni eficiente, por lo cual necesitaremos instalar un IDE, que no es más que un editor de texto con mejores funcionalidades.

En tal sentido, existen varios IDEs tales como Visual Studio Code, PyCharm, Jupyter Notebook, Jupyter Lab, Atom, Spyder. En el presente curso, se recomienda optar por Visual Studio Code o PyCharm (el usuario es libre de probar e interactuar con diferentes IDEs y elegir el que le resulte más cómodo)
