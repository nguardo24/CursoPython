# Entornos virtuales

Previamente a hablar de los conceptos de módulo, librería y paquete en `Python`, es bueno entender qué son y para qué sirven los entornos virtuales.

Instalar paquetes de `Python` puede tornarse engorroso, debido a que las dependencias entre paquetes y en librerías externas puede ser dificil de gestionar. Muchas veces, una herramienta específica puede requerir una versión más vieja de un paquete o incluso, una versión más vieja de `Python`. Además, siempre se considera como buena práctica, instalar paquetes desde un solo recurso o fuente principalmente, como por ejemplo, el mismo canal de conda.

En términos generales, un entorno de `Python` (*Python environment*) es una instalación local separada de todas las herramientas requeridas para una tarea particular, incluidos el intérprete de Python, los paquetes y las librerías subyacentes.

# Cómo crear un entorno virtual

`Python` contiene herramientas incorporadas y oficiales para gestionar entornos: el módulo `venv` y el paquete `virtualenv`. Adicionalmente, también puede utilizarse `conda` (gestor de paquetes del cuál hablaremos más adelante) para manejar los entornos de `Python`.

En tal sentido, y siempre trabajando desde una línea de comandos, se puede crear un entorno virtual.

## Módulo `venv`

Dentro de un entorno de lineas de comando, y estando ubicado en la carpeta correspondiente, se debe escribir el siguiente comando:

`python -m venv [ENV_NAME]` donde ENV_NAME representa el nombre del entorno

## Gestor `conda`

Dentro de un entorno de lineas de comando, y estando ubicado en la carpeta correspondiente, se debe escribir el siguiente comando:

`conda create --name [ENV_NAME]` donde ENV_NAME representa el nombre del entorno
