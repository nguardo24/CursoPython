# Instalación de Python

Para poder empezar a escribir código en Python, el primer paso consiste en instalar el intérprete.

En tal sentido, se puede instalar Python de varias formas:

1. Descargando el intérprete desde la página oficial
2. Descargando la distribución Anaconda
3. Descargar la distribución miniconda (versión minimalista de Anaconda)

## 1. Descargar intérprete de Python desde la pagina oficial e instalar de manera local

Dirigirse a la pagína [Descargar Python](https://www.python.org/downloads/) y descargar el archivo ejecutable (.exe) para dar inicio a la instalación (notar que la ultima versión de Python a la fecha es la 3.12.2)



**IMPORTANTE:** Al momento de instalar Python, se debe agregar el mismo como variable de entorno (Add python.exe to PATH).



## 2. Descargar e instalar Anaconda Distribution

Dirigirse a la pagina [Anaconda Download](https://www.anaconda.com/download), descargar el archivo ejecutable (.exe) e instalar



## 3. Descargar e instalar miniconda
Dirigirse a la página [Miniconda Install](https://docs.anaconda.com/free/miniconda/miniconda-install/) y seguir los pasos de instalación



# Instalar entorno de desarrollo integrado (Integrated Development Environment - IDE)

Una vez que tenemos `Python` instalado en el sistema, podríamos comenzar a escribir código, utilizando por ejemplo la terminal del sistema o el IDLE de Python.

Sin embargo, esta opción no es la más cómoda ni eficiente, por lo cual necesitaremos instalar un IDE, que no es más que un editor de texto con mejores funcionalidades para poder escribir y crear *scripts*. En tal sentido, existen varios IDEs tales como Visual Studio Code, PyCharm, Jupyter Notebook, Jupyter Lab, Atom y Spyder, entre otros. En el presente curso, se recomienda optar por Visual Studio Code o PyCharm (el usuario es libre de probar e interactuar con diferentes IDEs y elegir el que le resulte más cómodo).

En el caso de Visual Studio Code, dirigirse a la página [Download Visual Studio Code](https://code.visualstudio.com/download) y descargar el instalador correspondiente de acuerdo con el sistema operativo (Linux, Windows o MacOS). Una vez descargado, dar inicio a la instalación, corriendo el mismo. Adicionalmente, si el/la usuario/a desea, puede realizar la instalación del programa utilizando línea de comandos (Command Line Interface - CLI)

En el caso de PyCharm, dirigirse a la página [Descargar PyCharm](https://www.jetbrains.com/es-es/pycharm/download/) para dar inicio a la descarga del instalador. En este caso, la página se adecuará de acuerdo con el sistema operativo del usuario. Una vez descargado, dar inicio a la instalación del mismo.

PyCharm se presenta en dos ediciones: Community (gratis) y Professional (paga). A los fines del curso, se recomienda optar por la primera.
